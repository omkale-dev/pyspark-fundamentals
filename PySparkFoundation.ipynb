```python
from pyspark.sql.functions import *
data=[('Om','16-08-1999'),('Kartik','28-02-2002'),('Shashwat','21-02-2000')]
schema=['name','dob']
df=spark.createDataFrame(data,schema)
df.withColumn('Today',current_date())\
.withColumn('dob',to_date(df.dob,'dd-MM-yyyy'))\
.withColumn('changeFormat',date_format('dob','dd/MM/yyyy'))\
.show()
```

```python
from pyspark.sql.types import *
@udf(FloatType())
def days_to_years(days):
    return days/365
@udf(StringType())
def string_day(num):
    days=['Mon','Tue','Wed','Thurs','Fri','Sat','Sun']
    return days[num-1]

#  months/days give - value to substract
df.withColumn('age_days',datediff(current_date(),to_date(df.dob,'dd-MM-yyyy')))\
.withColumn('age_years',days_to_years('age_days'))\
.withColumn('months_between',months_between(current_date(),to_date(df.dob,'dd-MM-yyyy')))\
.withColumn('add_months',add_months(current_date(),12))\
.withColumn('add_days',date_add(current_date(),1))\
.withColumn('year',year(current_date()))\
.withColumn('month',month(current_date()))\
.withColumn('today',string_day(dayofweek(current_date())))\
.show()
```

```python
# suppose we want top sal for all dept then we can use this
from pyspark.sql.functions import *
from pyspark.sql.window import Window
data=[
    (1,'Om','M',2000,'IT'),
    (2,'Kartik','F',3000,'IT'),
    (2,'Kart','F',3000,'Sales'),
    (3,'Shas','M',4300,'Ops'),
    (4,'Shashi','F',4300,'Ops'),
    (5,'Kartey','F',3000,'HR'),
    (6,'Shusma','F',4000,'Sales'),
    (7,'Sunny','M',4500,'Ops'),
    (8,'Shilpa','F',300,'Ops')
    
]
schema=['id','name','gender','salary','dept']
df=spark.createDataFrame(data,schema)
window=Window.partitionBy('dept').orderBy(col('salary').desc())
df1=df.withColumn('row_number',row_number().over(window))
df2=df.withColumn('rank',rank().over(window))
df3=df.withColumn('dense_rank',dense_rank().over(window))
df1.show()
# df2.show()
# df3.show()
# highest salary from all dept
df1.select("*").filter(df1.row_number==1).show()
```
